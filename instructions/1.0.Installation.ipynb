# Anaconda python

Anaconda python is installed in the main account "cavelab". Pylidar v3.8.12 and conda v4.10.3. The python path needs to be added to your ~/.bashrc profile. Don't forget to source ~/.bashrc:

# Pylidar

```python
%%bash
conda config --add channels conda-forge
conda config --add channels rios
conda create -n vpp pylidar # vpp is the name of your environment - change if you want
source activate vpp # omit 'source' on Windows
conda install pynninterp
conda config --set channel_priority strict #see https://groups.google.com/g/pylidar/c/O6G1lE-9dqo
conda update pylidar
conda install tbb=2020.2
```

Now download the RIEGL libraries from www.riegl.com (members area):

RiVLib 2.5.10 x86 64bit Linux gcc9
RiWaveLib WFM 2.5.10 x86 64bit Linux gcc9

and put in a folder yourrepo/RIEGL

and add the libraries to your LD_LIBRARY_PATH in your ~/.bashrc. Example:

```python
export LD_LIBRARY_PATH=$LD_LIBRARY_PATH:/home/uceskca/code/plantprofiles/RIEGL/riwavelib_wfm-2_5_10-x86_64-linux-gcc9/lib
export LD_LIBRARY_PATH=$LD_LIBRARY_PATH:/home/uceskca/code/plantprofiles/RIEGL/rivlib-2_5_10-x86_64-linux-gcc9
```

Now test if the installation was successfull. Download the latest testdata at https://github.com/ubarsc/pylidar/wiki/Testdata and:

```python
%%bash
cd ~/Downloads
source activate vpp
pylidar_test -i testdata_10.tar.bz2 -p /tmp
```
